```python
from manim import *
import numpy as np 

class ComplexNumbers(Scene):
    def construct(self):
        
#         title:
        title = Tex(r"Kompleks sanlar")
        title.to_corner(UP)
        self.play(Write(title))
        
        
        # Define each line as a separate MathTex
        line1 = MathTex(r"a z^2+b z+c=0", substrings_to_isolate="z")
        line2 = MathTex(r"z=\frac{-b \pm \sqrt{b^2-4ac}}{2a}", substrings_to_isolate="z")
        line3 = MathTex(r"D=b^2-4ac<0")
        line4 = Tex(r"Çözüwi Ýok ???")
        
        # Set color for "z" in each line
        line1.set_color_by_tex("z", YELLOW)
        line2.set_color_by_tex("z", YELLOW)
        
        # Position lines one below the other
        line1.move_to(UP)
        line2.next_to(line1, DOWN)
        line3.next_to(line2, DOWN)
        line4.next_to(line3, DOWN)
        
        # Animate each line separately with Write and wait in between
        self.play(Write(line1))
        self.wait(1)  # Wait a second before the next line
        self.play(Write(line2))
        self.wait(1)
        self.play(Write(line3))
        self.wait(2)
        self.play(Write(line4))
        self.wait(2)
        f_group=VGroup(line1, line2, line3, line4)
        
        self.play(f_group.animate.move_to(LEFT*3.5))
        self.wait(1)
        
        line10 = MathTex(r"z^2-6z+13=0", substrings_to_isolate="z")
        line20 = MathTex(r"z=\frac{6 \pm \sqrt{6^2-4\cdot 13 }}{2}", substrings_to_isolate="z")
        line30 = MathTex(r"z=\frac{6 \pm \sqrt{36-52}}{2}", substrings_to_isolate="z")
        line31 = MathTex(r"z=\frac{6 \pm \sqrt{-16}}{2}", substrings_to_isolate="z")
        lineI = MathTex(r"\sqrt{-1}=i")
        line32 = MathTex(r"z=\frac{6 \pm 4 i }{2}=3 \pm 2 i", substrings_to_isolate="z")
        
        line40 = MathTex(r"z=","3"," + ","2", " i", substrings_to_isolate="z")
        line41 = MathTex(r"z=","3"," - ","2", " i", substrings_to_isolate="z")
        
        # Set color for "z" in each line
        line10.set_color_by_tex("z", YELLOW)
        line20.set_color_by_tex("z", YELLOW)
        line30.set_color_by_tex("z", YELLOW)
        line31.set_color_by_tex("z", YELLOW)
        line32.set_color_by_tex("z", YELLOW)
        line40.set_color_by_tex("z", YELLOW)
        line41.set_color_by_tex("z", YELLOW)

        # Position lines one below the other
        line10.move_to(UP*2+RIGHT*1.8)
        line20.next_to(line10, DOWN)
        line30.move_to(line20)  # Align line30 with line20
        line31.move_to(line20)  # Align line31 with line20
        lineI.next_to(line31, DOWN)
        line32.next_to(lineI, DOWN)
        line40.next_to(line32, DOWN)
        line41.next_to(line40, DOWN)
        
        e_group=VGroup(line10, line31, line32, line41)
        
        # animate 
        self.play(Write(line10))
        self.wait(1)
        self.play(Write(line20))
        self.wait(1)
        self.play(TransformMatchingTex(line20, line30))
        self.wait(1)
        self.play(TransformMatchingTex(line30, line31))
        self.wait(3)
        self.play(Write(lineI))
        self.wait(4)
        self.play(Transform(line31, line32))
        self.wait(3)
        self.play(Write(line40), Write(line41))
        self.wait(3)
        
        # Surround the i and answer 
        framebox1 = SurroundingRectangle(lineI, buff = .1)
        framebox2 = SurroundingRectangle(line40, buff = .2)
        
        self.play(Create(framebox1))
        self.wait(1)
        self.play(Create(framebox2))
        self.wait(1)
        
        lineI_f=VGroup(lineI, framebox1)
        line40_f=VGroup(line40, framebox2)
        
        #isolate the definition 
        self.play(Unwrite(f_group), Unwrite(e_group))
        self.play(lineI_f.animate.scale(1.4).move_to(UP*2+LEFT*3), line40_f.animate.scale(1.4).next_to(UP*2+RIGHT))
        self.wait(2)
        
        self.play(Uncreate(framebox1), Uncreate(framebox2))
        
        self.wait(3)
        
        real_write=MathTex(r"Re \{z \}=3", substrings_to_isolate="3").set_color_by_tex("3", '#00FFFF')
        im_write=MathTex(r"Im \{z\}= 2", substrings_to_isolate="2").set_color_by_tex("2", '#00FF00')
        real_write.next_to(line40_f, DOWN)
        im_write.next_to(real_write, DOWN)
        
        real_rec=SurroundingRectangle(line40[2], color='#00FFFF')
        im_rec=SurroundingRectangle(line40[4], color='#00FF00')
        self.play(Create(real_rec), Write(real_write))
        self.wait(2)
        self.play(ReplacementTransform(real_rec, im_rec), Write(im_write))
        self.wait(2)
        
        self.play(Uncreate(im_rec))
        self.wait(1)
        
        pair=MathTex(r"(3, 2)", substrings_to_isolate=["3", "2"]).scale(1.2)
        pair.set_color_by_tex("3", '#00FFFF')
        pair.set_color_by_tex("2", '#00FF00')
        pair.next_to(im_write, DOWN)
        
        self.play(Write(pair))
        self.wait()
        
#       Creating the axis:
        ax = Axes(x_range=[-4, 4, 1], y_range=[-4, 4, 1]).add_coordinates()
        ax_label=ax.get_axis_labels(
            MathTex("x, Re\{z\}").scale(0.7), MathTex("y, Im\{z\}").scale(0.9)
        )
        dot_axes = Dot(ax.coords_to_point(3, 2), color=RED)
        dot_label=MathTex(r"(3, 2)", substrings_to_isolate=["3", "2"]).next_to(dot_axes, UP)
        dot_label.set_color_by_tex("3", '#00FFFF')
        dot_label.set_color_by_tex("2", '#00FF00')
        
        lines = ax.get_lines_to_point(ax.c2p(3,2))
        
        ax_group=VGroup(ax, dot_axes,dot_label, ax_label, lines)
        ax_group.scale(0.8).move_to(LEFT*4)
        
#       Animate axes:
        self.play(Unwrite(lineI))
        self.wait(1)
        self.play(Create(ax), Create(ax_label))
        self.play(FadeIn(dot_axes), Create(lines))
        self.play(Transform(pair, dot_label))
        self.wait(3)
        
#       Tranforming for the general case: 
        line40_g=MathTex(r"z= x +i y ", substrings_to_isolate=["x", "y"]).scale(1.2).move_to(line40.get_center())
        line40_g.set_color_by_tex("x", '#00FFFF')
        line40_g.set_color_by_tex("y", '#00FF00')
        
        real_write_g=MathTex(r"Re \{z \}= x ", substrings_to_isolate="x").next_to(line40_g, DOWN).set_color_by_tex("x",'#00FFFF')
        im_write_g=MathTex(r"Im \{z\}= y ", substrings_to_isolate="y").next_to(real_write_g, DOWN).set_color_by_tex("y", '#00FF00')
        dot_label_g=MathTex(r"( x , y )", substrings_to_isolate=["x", "y"]).move_to(dot_label.get_center())
        dot_label_g.set_color_by_tex("y", '#00FF00')
        dot_label_g.set_color_by_tex("x", '#00FFFF')
        

        self.play(ReplacementTransform(line40, line40_g), 
                  ReplacementTransform(real_write, real_write_g), 
                  ReplacementTransform(im_write, im_write_g), 
                  ReplacementTransform(pair, dot_label_g))
        self.wait(3)
        label_ax=Tex("Kompleks tekizligi").scale(0.8).next_to(ax, DOWN)
        label_ax2=Tex("Argand tekizligi").scale(0.8).next_to(label_ax, DOWN)
        
        self.play(Write(label_ax))
        self.play(Write(label_ax2))
        self.wait(2)
        self.play(Unwrite(label_ax), Unwrite(label_ax2))
        self.wait()
        
        self.play(Unwrite(real_write_g), Unwrite(im_write_g))
        self.wait()
        
        text_pol=Tex("Polýar koordinatalar sistemasy").scale(0.8).next_to(line40_g, DOWN)
        pol_cor=MathTex(r"r",",","\theta").next_to(text_pol, DOWN)
        
        r_vector=Line(start=ax.c2p(0, 0), end=ax.c2p(3,2))
        r_label=MathTex("r").next_to(r_vector.get_center(), UP)

        
        self.play(Write(text_pol))
        self.play(GrowFromPoint(r_vector, point=r_vector.get_start()), Write(r_label), run_time=2)
        
        self.wait()
#         adding angle:
        angle_arc = Angle(
            Line(ax.c2p(0, 0), ax.c2p(1, 0)),  # x-axis as reference line
            r_vector,
            radius=1.3,
            other_angle=False,
            color=YELLOW
        )
    
        angle_label=MathTex(r"\theta", color=YELLOW)
        angle_label.next_to(angle_arc, RIGHT, buff=0.2)

        # Animate drawing of the angle arc and label
        self.play(Create(angle_arc), run_time=2)
        self.play(FadeIn(angle_label))
        self.wait()
        
        pol_r=MathTex(r"(r,").next_to(text_pol, DOWN)
        pol_angle=MathTex(r"\theta )", color=YELLOW).next_to(pol_r, RIGHT)
        
        self.play(
            TransformFromCopy(r_label, pol_r),  # Copy "r" into combined label
            TransformFromCopy(angle_label, pol_angle),   # Copy "\\theta" into combined label
            run_time=3
        )
        self.wait(1)
        
#         writing relation 
        
        
        x_r=MathTex(r"x=r \cos{\theta}", substrings_to_isolate=["x", r"\theta"]).next_to(pol_r, DOWN)
        x_r.set_color_by_tex("x",'#00FFFF')
        x_r.set_color_by_tex(r"\theta", YELLOW)
        y_r=MathTex(r"y=r \sin{\theta}", substrings_to_isolate=["y", r"\theta"]).next_to(x_r, DOWN)
        y_r.set_color_by_tex("y",'#00FF00')
        y_r.set_color_by_tex(r"\theta",YELLOW)
        
        r_mod=MathTex(r"r=\sqrt{x^{2}+y^{2}}").next_to(y_r, DOWN)
        
        r_mod_label=Tex("moduly:").scale(0.8).next_to(r_mod, LEFT)
        
        self.play(
            Write(x_r),  
            Write(y_r),
            run_time=3
        )
        self.wait(1)
        self.play(Write(r_mod))
        self.wait(1)
        self.play(Write(r_mod_label))
        
        line50=MathTex(r"z=x+iy=r \cos \theta+ i r \sin \theta", substrings_to_isolate=["x", "y", "\theta"]).next_to(r_mod, DOWN)
        line50.set_color_by_tex("x",'#00FFFF')
        line50.set_color_by_tex("y",'#00FF00')
        line50.set_color_by_tex("\theta", YELLOW)
        
        self.play(TransformFromCopy(line40_g, line50))     
        self.wait(3)
        
        #1.11.2024
        line51=MathTex(r"z=x+iy=r(\cos \theta+ i \sin \theta)",  substrings_to_isolate=["x", "y", r"\theta"]).next_to(r_mod, DOWN)
        line51.set_color_by_tex("x", '#00FFFF')
        line51.set_color_by_tex("y",'#00FF00')
        line51.set_color_by_tex(r"\theta", YELLOW)
        self.play(TransformMatchingTex(line50, line51))
        self.wait(3)
        
        last_group=VGroup(line40_g, text_pol, pol_cor, x_r, y_r, r_mod, r_mod_label, pol_r, pol_angle)
        pol_group=VGroup(r_vector, r_vector, r_label, angle_arc, angle_label)
        
        self.play(FadeOut(last_group), FadeOut(ax_group), FadeOut(pol_group), line51.animate.move_to(title.get_center()+DOWN), )
        self.wait(2)
        
%manim -qm -v WARNING ComplexNumbers
```

```python
from manim import *
import numpy as np

class AnimateComplexNumber(Scene):
    def construct(self):
        # Function to calculate polar coordinates
        def calculate_r_theta(x, y):
            r = np.sqrt(x**2 + y**2)
            theta = np.arctan2(y, x)
            return r, theta
        
        # Initial values
        x_tracker = ValueTracker(2)
        y_tracker = ValueTracker(1)
        
        # Draw the axes
        ax = Axes(x_range=[-5, 5, 1], y_range=[-5, 5, 1]).add_coordinates()
        ax_label = ax.get_axis_labels(
            MathTex("x, Re\\{z\\}").scale(0.7), MathTex("y, Im\\{z\\}").scale(0.5)
        )
        
        # Initial dot at the complex point
        dot_axes = Dot(color=RED)
        dot_axes.add_updater(lambda mob: mob.move_to(ax.c2p(x_tracker.get_value(), y_tracker.get_value())))
        
        # Lines from axes to point
        lines = ax.get_lines_to_point(ax.c2p(x_tracker.get_value(), y_tracker.get_value()))
        lines.add_updater(lambda mob: mob.become(ax.get_lines_to_point(ax.c2p(x_tracker.get_value(), y_tracker.get_value()))))
        
        # Polar coordinates labels and vector
        r_vector = Line(start=ax.c2p(0, 0), end=ax.c2p(x_tracker.get_value(), y_tracker.get_value()))
        r_vector.add_updater(lambda mob: mob.put_start_and_end_on(ax.c2p(0, 0), ax.c2p(x_tracker.get_value(), y_tracker.get_value())))

        # Angle arc and label
        angle_arc = Angle(
            Line(ax.c2p(0, 0), ax.c2p(1, 0)),
            r_vector,
            radius=1.3,
            color=YELLOW
        )
        angle_arc.add_updater(lambda mob: mob.become(Angle(
            Line(ax.c2p(0, 0), ax.c2p(1, 0)),
            r_vector,
            radius=1,
            color=YELLOW
        )))

        # Update labels for polar and rectangular forms
        def get_dot_label():
            x, y = x_tracker.get_value(), y_tracker.get_value()
            r, t = calculate_r_theta(x, y)
            label = MathTex(
                fr"({x:.1f}, {y:.1f}) \\",
                fr"{x:.1f}+i{y:.1f} \\",
                fr"{r:.1f}(\cos{t:.1f}+i \sin{t:.1f}) \\",
                fr"{r:.1f}e^{{i{t:.1f}}}"
            ).scale(0.5).next_to(dot_axes, UP)
            label.set_color_by_tex(fr"{x:.1f}", '#00FFFF')
            label.set_color_by_tex(fr"{y:.1f}", '#00FF00')
            label.set_color_by_tex(fr"{t:.1f}", YELLOW)
            label.set_color_by_tex(fr"{r:.1f}", WHITE)
            return label

        dot_label = get_dot_label()
        dot_label.add_updater(lambda mob: mob.become(get_dot_label()))

        # Radius (magnitude) label
        r_label = MathTex(fr"r={calculate_r_theta(x_tracker.get_value(), y_tracker.get_value())[0]:.1f}").scale(0.5)
        r_label.add_updater(lambda mob: mob.next_to(r_vector.get_center(), UP).become(MathTex(
            fr"r={calculate_r_theta(x_tracker.get_value(), y_tracker.get_value())[0]:.1f}").scale(0.5)))

        # Angle label
        angle_label = MathTex(
            fr"\theta={calculate_r_theta(x_tracker.get_value(), y_tracker.get_value())[1]:.1f}", color=YELLOW
        ).scale(0.5).next_to(angle_arc, RIGHT)
        angle_label.add_updater(lambda mob: mob.next_to(angle_arc, RIGHT, buff=0.2).become(MathTex(
            fr"\theta={calculate_r_theta(x_tracker.get_value(), y_tracker.get_value())[1]:.1f}", color=YELLOW
        ).scale(0.5).next_to(angle_arc, RIGHT)))

        # Add objects to the scene
        self.add(ax, ax_label, dot_axes, lines, dot_label, r_vector, angle_arc, r_label, angle_label)

        # Animate movement of point and labels by updating x_tracker and y_tracker
        self.play(x_tracker.animate.set_value(3), y_tracker.animate.set_value(4), run_time=2)
        self.play(x_tracker.animate.set_value(2), y_tracker.animate.set_value(1), run_time=2)
        self.play(x_tracker.animate.set_value(1), y_tracker.animate.set_value(2), run_time=2)
        self.play(x_tracker.animate.set_value(-2), y_tracker.animate.set_value(3), run_time=2)
        self.play(x_tracker.animate.set_value(-3), y_tracker.animate.set_value(0), run_time=2)
#         self.play(x_tracker.animate.set_value(-1), y_tracker.animate.set_value(-2), run_time=2)
#         self.play(x_tracker.animate.set_value(0), y_tracker.animate.set_value(-3), run_time=2)
#         self.play(x_tracker.animate.set_value(2), y_tracker.animate.set_value(-2), run_time=2)
        
        
        
        
        # Clean up
#         dot_axes.clear_updaters()

%manim -qm -v WARNING AnimateComplexNumber
```

```python
from manim import *
import numpy as np

class Pendulum(Scene):
    def construct(self):
#         constants:

        time=ValueTracker(0)
        theta_max=30/180*PI 
        l=4
        g=10
        w=np.sqrt(g/l)
        T=2*PI/w
        
#         Setting desired pendulum position:
        p_x=0
        p_y=3 
        shift_req=p_x*RIGHT+p_y*UP
        
    
#         setting theta 
        theta=DecimalNumber().set_color(BLACK).move_to(5*RIGHT)
        theta.add_updater(lambda m: m.set_value((theta_max)*np.sin(w*time.get_value())))
        
        self.add(theta)
        
#         make a line:
        def get_line(x, y):
            line_here=Line(start=ORIGIN+shift_req, end=x*RIGHT+y*UP+shift_req, color=GREY)
            
            global line_vertical
            line_vertical=DashedLine(start=line_here.get_start(), end=line_here.get_start()+3*DOWN, color=GREY)
            
            return line_here
        
        line=always_redraw(lambda: get_line(l*np.sin(theta.get_value()), -l*np.cos(theta.get_value())))
        
        self.add(line)
        self.add(line_vertical)
        
#         creating an arc:
        def angle_arc(theta):
            global angle 
            global arc_text
            
            if theta==0:
                angle=VectorizedPoint().move_to(10*RIGHT)           # moving out of screen 
                arc_text=VectorizedPoint().move_to(10*RIGHT)
            else:
                if theta>0:
                    angle=Angle(line, line_vertical, quadrant=(1, 1), other_angle=True, color=YELLOW, fill_opacity=0)
                    
                elif theta<0:
                    angle=Angle(line, line_vertical, quadrant=(1, 1), other_angle=False, color=YELLOW, fill_opacity=0)
                    
            
            return angle
        
        angle=always_redraw(lambda: angle_arc(theta.get_value()))
        
        self.add(angle)
        
        arc_text=Tex(r"$\theta$").scale(0.7)
        
        arc_text.add_updater(lambda m: m.next_to(angle, DOWN))
        self.add(arc_text)
        
#         making a bob
        def get_ball(x, y):
            dot=Dot(fill_color=BLUE, fill_opacity=1).move_to(x*RIGHT+y*UP+shift_req).scale(l)
            return dot 
        
        ball=always_redraw(lambda: get_ball(l*np.sin(theta.get_value()), -l*np.cos(theta.get_value())))
        
        self.add(ball)
        
        self.play(time.animate.set_value(3*T), rate_func=linear, run_time=3*T)

%manim -qm -v WARNING Pendulum
```

```python
from manim import *
import numpy as np

class WaveFunctionEvolution(Scene):
    def construct(self):
        # Initialize time as an attribute of the class
        self.time = 0

        # Parameters for the wave packet
        A = 1      # Amplitude
        x0 = -2    # Initial position
        sigma = 0.5  # Width of the Gaussian
        k = 2     # Wave number
        m = 1     # Mass
        hbar = 1  # Reduced Planck's constant
        omega = hbar * k**2 / (2 * m)  # Angular frequency

        # Function to calculate the real part of Psi
        def psi_real(x, t):
            return A * np.exp(-(x - x0)**2 / (4 * sigma**2)) * np.cos(k * x - omega * t)

        # Axes for the plot
        axes = Axes(
            x_range=[-5, 5, 1],
            y_range=[-1, 1, 0.5],
            axis_config={"color": BLUE},
        )

        # Labels for axes (using simple text to avoid LaTeX issues)
        labels = axes.get_axis_labels(x_label="x", y_label=r"Re(\Psi)")

        # Initial graph
        wave_function = always_redraw(lambda: axes.plot(
            lambda x: psi_real(x, self.time),
            color=YELLOW,
        ))

        # Add everything to the scene
        self.add(axes, labels, wave_function)

        # Animate the wave function over time
        self.play(Create(axes), Write(labels))
        self.wait(1)
        
        # Run the animation for the wave function evolving over time
        def update_wave_function(mob, dt):
            self.time += dt  # Increment time
            mob.become(axes.plot(lambda x: psi_real(x, self.time), color=YELLOW))

        wave_function.add_updater(update_wave_function)
        self.wait(10)  # Let it run for 10 seconds
        wave_function.remove_updater(update_wave_function)

        
%manim -qm -v WARNING WaveFunctionEvolution
```

```python
from manim import *
import numpy as np

class AlternatingCircuit(Scene):
    def construct(self):
        # Initialize time as an attribute of the class
        self.time = 0

        # 1. Draw the Circuit Components (Top)
        
        # Position the components
        top_y = 2.5  # Y position for the circuit components
        start_point = LEFT * 4 + UP * top_y
        right_point = RIGHT * 4 + UP * top_y
        
        # Define the components
        inductor = Square(side_length=0.5, color=BLUE).shift(LEFT * 2 + UP * top_y)  # represents L
        capacitor = VGroup(
            Line(UP * 0.25, DOWN * 0.25, color=RED), 
            Line(UP * 0.25, DOWN * 0.25, color=RED).next_to(RIGHT * 0.2, buff=0)
        ).shift(RIGHT * 2 + UP * top_y)  # represents C

        # Labels for components
        inductor_label = Text("L", font_size=24).next_to(inductor, UP)
        capacitor_label = Text("C", font_size=24).next_to(capacitor, UP)

        # Draw circuit connections
        circuit_path = VGroup(
            Line(start_point, inductor.get_left()),    # left to inductor
            Line(inductor.get_right(), LEFT * 1 + UP * top_y),   # inductor to center left
            Line(LEFT * 1 + UP * top_y, RIGHT * 1 + UP * top_y), # center connection
            Line(RIGHT * 1 + UP * top_y, capacitor.get_left()),  # center right to capacitor
            Line(capacitor.get_right(), right_point),  # capacitor to right side
        )

        # Add components and labels to the scene
        self.play(Create(circuit_path), Create(inductor), Create(capacitor))
        self.play(Write(inductor_label), Write(capacitor_label))

        # 2. Animate the Sinusoidal Wave (Bottom)

        # Axes for the waveform
        axes = Axes(
            x_range=[0, 4 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            x_length=8,
            y_length=2,
            axis_config={"color": BLUE}
        ).shift(DOWN * 2)

        # Labels for axes
        axes_labels = axes.get_axis_labels(x_label="Time", y_label="Amplitude")

        # Sinusoidal function to represent the oscillation
        sinusoid = always_redraw(lambda: axes.plot(
            lambda x: np.sin(x - self.time),
            color=YELLOW
        ))

        # Add axes and sinusoidal wave to the scene
        self.add(axes, axes_labels, sinusoid)

        # Animate the sinusoidal wave over time
        def update_time(dt):
            self.time += dt  # Update time for oscillation

        # Run the sinusoidal animation
        self.add_updater(update_time)
        self.wait(10)  # Let it animate for 10 seconds
        self.remove_updater(update_time)


%manim -qm -v WARNING AlternatingCircuit
```
